## 画像ダウンロード

### Subtask:
提供されたURLから画像をダウンロードし、`pig_image` ディレクトリに保存します。ファイル名は `pig_01.jpg` から `pig_11.jpg` とします。`pig_09` のURLは `https://www.shutterstock.com/image-photo/adorable-typical-black-white-young-600w-2587269077.jpg` を使用します。


**Reasoning**:
Import the `requests` library and download the images from the provided URLs, saving them to the 'pig_image' directory with the specified filenames, using the alternative URL for pig_09.



```python
import requests

urls = [
    "https://i.ytimg.com/vi/kSytjGSAeT4/maxresdefault.jpg",
    "https://up.gc-img.net/post_img/2024/07/O3RnYMRAWVEU4zQ_ZS3gK_275.jpeg",
    "https://i.pinimg.com/736x/0b/e5/8f/0be58f6133c495e0ca1a565285f06164.jpg",
    "https://p.potaufeu.asahi.com/31d9-p/picture/27241884/66194ac0da3c31eaf207e74d94842a60.jpg",
    "https://research.image.itmedia.co.jp/wp-content/uploads/2023/04/1682573327_20230427_t_02.jpg",
    "https://www.pfirst.jp/on/demandware.static/-/Sites-pfirst-Library/ja_JP/dwd031f7f1/dogbook_top02.jpg",
    "https://maruone.com/images/upload/2023/02/2cb834b33c7076f72863fc49710abb2f.jpg",
    "https://d1d7kfcb5oumx0.cloudfront.net/articles/images/643bbbd36db64c07d73a75b5/slide_1681636304956.jpg",
    "https://nuena.com/wp-content/uploads/2013/03/buddy.jpg",
    "https://up.gc-img.net/post_img/2024/07/O3RnYMRAWVEU4zQ_p81uu_288.jpeg",
    "https://review-shopping.c.yimg.jp/images/yokadoh-shop_c039/1421a5b0-38ad-4e44-ae8b-b3aa7156bf52.jpeg"
]

alternative_url_pig09 = "https://www.shutterstock.com/image-photo/adorable-typical-black-white-young-600w-2587269077.jpg"

for i, url in enumerate(urls):
    filename = f'pig_image/pig_{i+1:02d}.jpg'
    if i == 8: # Index 8 corresponds to pig_09
        url_to_download = alternative_url_pig09
    else:
        url_to_download = url

    try:
        response = requests.get(url_to_download)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {filename}")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {url_to_download}: {e}")
```

## ローカルリンクリスト作成

### Subtask:
`pig_image` ディレクトリに保存された画像へのローカルリンクをリストとして表示します。


**Reasoning**:
List the files in the 'pig_image' directory and create a list of local markdown links for each file.



```python
import os

image_dir = 'pig_image'
image_files = sorted(os.listdir(image_dir))
local_links = []
for file in image_files:
    if file.endswith('.jpg'):
        # Extract the base name without extension for the link text and replace "pig" with "dog"
        base_name = os.path.splitext(file)[0].replace('pig', 'dog')
        local_link = f"[{base_name}]({image_dir}/{file})" # The file name in the path is already "dog_xx.jpg"
        local_links.append(local_link)

for link in local_links:
    display(link)
```

## ディレクトリ内のファイル名の置換

### Subtask:
`dog_image` ディレクトリ内の画像ファイル名に含まれる "pig_" を "dog_" に置き換えます。

**Reasoning**:
Import the `os` module and iterate through the files in the 'dog_image' directory. For each file, if it contains "pig_" and is a file (not a directory), rename it by replacing "pig_" with "dog_".

```python
import os

image_dir = 'dog_image'

if os.path.exists(image_dir):
    for filename in os.listdir(image_dir):
        if 'pig_' in filename and os.path.isfile(os.path.join(image_dir, filename)):
            new_filename = filename.replace('pig_', 'dog_')
            old_filepath = os.path.join(image_dir, filename)
            new_filepath = os.path.join(image_dir, new_filename)
            os.rename(old_filepath, new_filepath)
            print(f"Renamed '{filename}' to '{new_filename}'")
else:
    print(f"Directory '{image_dir}' not found.")
```

## 画像の表示 (dog_06.jpg と dog_07.jpg)

### Subtask:
ユーザーが指摘した `dog_07.jpg` と、関連性の高い `dog_06.jpg` を表示します。

**Reasoning**:
Import the `Image` and `display` functions from `IPython.display` and the `os` module. Construct the file paths for 'dog_image/dog_06.jpg' and 'dog_image/dog_07.jpg' and display them using `display(Image(filename=file_path))`.

```python
from IPython.display import Image, display
import os

image_dir = 'dog_image'
image_files_to_display = ['dog_06.jpg', 'dog_07.jpg']

print("指定された画像を表示します:")
for file in image_files_to_display:
    file_path = os.path.join(image_dir, file)
    if os.path.exists(file_path):
        print(f"\n--- {file} ---")
        try:
            display(Image(filename=file_path))
        except Exception as e:
            print(f"Error displaying {file}: {e}")
    else:
        print(f"ファイルが見つかりません: {file_path}")
```
